```python
np.random.seed(1)
array_1 = [''.join(random.choices(string.ascii_letters, k = 15)) for _ in range(5000)]
np.random.seed(1)
array_2 = [''.join(random.choices(string.ascii_letters, k = 15)) for _ in range(10000)]
np.random.seed(1)
array_3 = [''.join(random.choices(string.ascii_letters, k = 15)) for _ in range(15000)]
np.random.seed(1)
array_4 = [''.join(random.choices(string.ascii_letters, k = 15)) for _ in range(20000)]
np.random.seed(1)
array_5 = [''.join(random.choices(string.ascii_letters, k = 15)) for _ in range(25000)]
rand_str15 = [array_1, array_2, array_3, array_4, array_5]
```

```python
import matplotlib.pyplot as plt
# using the textbook codes
def findSmallest(arr):
    smallest = arr[0]
    smallest_index = 0
    for i in range(1, len(arr)):
        if arr[i] < smallest:
            smallest = arr[i]
            smallest_index = i
        
    return [smallest, smallest_index]

def selectionSort(arr):
    newArr = []
    start_time = time.time()
    for i in range(len(arr)):
        smallest = findSmallest(arr)[0]
        smallest_index = findSmallest(arr)[1]
        arr = np.delete(arr, smallest_index, axis=0)
        newArr.append(smallest)
    end_time = time.time()
    return end_time - start_time

def get_sort_value(ls_array):
    # loop through the list of array and return a list of 2 lists that has the lenght and the processing time
    results = []
    ls = [0, 1, 2, 3, 4]
    for i in ls:
        time_val = selectionSort(ls_array[i])
        lenght = len(ls_array[i])
        results.append([str(lenght), time_val*1000])
    return results

def plot_result(res_list):
    data = pd.DataFrame(res_list, columns=["Array_Size", "Time_select_sort"])
    data.plot(y='Time_select_sort', x='Array_Size')
    plt.show()
    
```

```python
result_int = get_sort_value(rand_int)
plot_result(result_int)
```

### executive summary
Selection sort is used to sort an array by repeatedly finding the smallest value in an unsorted array and put it at the beginning of the array. It has a runtime value of O(N^2) which is not very good when it comes time to sort a large number of data; we can see here, the more data there are for sorting the more time it will take.
Looking at the results, which are very surprising, there isn't much a difference in runtime between the following data types (integers, floats and text string of length 15). The surprising result was for text string of length 5. My guess when running this algorithm was for the text string of length 15 to take much longer to run. it seems like for text string of length 15, in addition to going through each cell in the array, the algorithm didn't have to do much work looking through each character in the cell to see which position the should be in. it might have found the answer way earlier. For instance:
aoasdhoajsdkas, boasdhoajsdkas, coasdhoajsdkas it will compare using the first character of the string. This hasn't been the case for text string of length 5.
